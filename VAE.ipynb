```python
import keras
from matplotlib import pyplot as plt
import tensorflow
from keras.layers import Input,Conv2D,MaxPooling2D,Conv2DTranspose,Reshape,Lambda
from keras.models import Model
from keras.optimizers import Adam
from keras.layers import concatenate
%matplotlib inline
```

```python
epochs = 100
inChannel = 3
x, y = 256,192
input_img = Input(shape = (x, y, inChannel))
latent_dim=2
```

```python
from keras import backend as K

def conv2d(input_img,n_filters,size=(3,3),batchnorm=True):
    conv=Conv2D(filters=n_filters,kernel_size=size,kernel_initializer='he_normal',padding='same')(input_img)
    if batchnorm:
        conv=keras.layers.BatchNormalization()(conv)
    conv=keras.layers.Activation('relu')(conv)
    
    conv=Conv2D(filters=n_filters,kernel_size=size,kernel_initializer='he_normal',padding='same')(input_img)
    if batchnorm:
        conv=keras.layers.BatchNormalization()(conv)
    conv=keras.layers.Activation('relu')(conv)
    return conv

def sampling(args):
    z_mean,z_log_var=args
    batch=K.shape(z_mean)[0]
    dim=K.int_shape(z_mean)[1]
    eps=K.random_normal(shape=(batch,dim))
    return z_mean+K.exp(0.5*z_log_var)*eps
#dropout=0.1
#batchnorm=True
#n_filters=16
def modified_variational_autoencoder(input_img,n_filters=16,dropout=0.1,batchnorm=True):
    #Encoder
    conv1=conv2d(input_img,n_filters*1,size=(3,3),batchnorm=batchnorm)
    pool1=MaxPooling2D((2,2))(conv1)
    pool1=keras.layers.Dropout(dropout)(pool1)

    conv2=conv2d(pool1,n_filters*2,size=(3,3),batchnorm=batchnorm)
    pool2=MaxPooling2D((2,2))(conv2)
    pool2=keras.layers.Dropout(dropout)(pool2)

    conv3=conv2d(pool2,n_filters*4,size=(3,3),batchnorm=batchnorm)
    pool3=MaxPooling2D((2,2))(conv3)
    pool3=keras.layers.Dropout(dropout)(pool3)

    conv4=conv2d(pool3,n_filters*8,size=(3,3),batchnorm=batchnorm)
    print("conv4 shape is",conv4.shape)
    pool4=MaxPooling2D((2,2))(conv4)
    pool4=keras.layers.Dropout(dropout)(pool4)

    conv5=conv2d(pool4,n_filters*16,size=(3,3),batchnorm=batchnorm)
    print(conv5.shape)
    pool5=MaxPooling2D((2,2))(conv5)
    pool5=keras.layers.Dropout(dropout)(pool5)
    
    conv6=conv2d(pool5,n_filters*32,size=(3,3),batchnorm=batchnorm)
    print(conv6.shape)
    shape=K.int_shape(conv6)

    flattened=keras.layers.Flatten()(conv6)
    print('Flattened shape is',flattened.shape)
    densed=keras.layers.Dense(256,activation='relu')(flattened)
    print("Densed shape is ",densed.shape)
    z_mean=keras.layers.Dense(latent_dim,name='z_mean')(densed)
    z_log_var=keras.layers.Dense(latent_dim,name='z_log_var')(densed)

    z=Lambda(sampling,output_shape=(latent_dim,),name='z')([z_mean,z_log_var])
    encoder=Model(input_img,[z_mean,z_log_var,z],name='encoder')

    #latent_input=Input(shape=(latent_dim,),name='z_sampling')
    #x=keras.layers.Dense(shape[1]*shape[2]*shape[3],activation='relu')(latent_input)
    x=keras.layers.Dense(shape[1]*shape[2]*shape[3],activation='relu')(z)
    x=Reshape((shape[1],shape[2],shape[3]))(x)

    #Decoder
    up7=Conv2DTranspose(n_filters * 16, (3, 3), strides = (2, 2), padding = 'same')(x)
    print(up7.shape,conv5.shape)
    up7=concatenate([up7,conv5])
    print(up7.shape)
    up7=keras.layers.Dropout(dropout)(up7)
    conv7=conv2d(up7,n_filters*16,size=(3,3),batchnorm=batchnorm)

    up8=Conv2DTranspose(n_filters * 8, (3, 3), strides = (2, 2), padding = 'same')(conv7)
    print(up8.shape)
    print(conv4.shape)
    up8=concatenate([up8,conv4])
    up8=keras.layers.Dropout(dropout)(up8)
    conv8=conv2d(up8,n_filters*4,size=(3,3),batchnorm=batchnorm)
    
    up9=Conv2DTranspose(n_filters * 4, (3, 3), strides = (2, 2), padding = 'same')(conv8)
    up9=concatenate([up9,conv3])
    up9=keras.layers.Dropout(dropout)(up9)
    conv9=conv2d(up9,n_filters*2,size=(3,3),batchnorm=batchnorm)

    up10=Conv2DTranspose(n_filters * 2, (3, 3), strides = (2, 2), padding = 'same')(conv9)
    up10=concatenate([up10,conv2])
    up10=keras.layers.Dropout(dropout)(up10)
    conv10=conv2d(up10,n_filters*1,size=(3,3),batchnorm=batchnorm)
    
    up11=Conv2DTranspose(n_filters * 1, (3, 3), strides = (2, 2), padding = 'same')(conv10)
    up11=concatenate([up11,conv1])
    up11=keras.layers.Dropout(dropout)(up11)
    conv11=conv2d(up11,n_filters*1,size=(3,3),batchnorm=batchnorm)

    output=Conv2D(3,(1,1),activation='sigmoid')(conv11)
    #decoder=Model(latent_input,output,name='decoder')
    #output1=decoder(encoder(input_img)[2])

    vae_model=Model(input_img,output,name='vae_model')

    
    return vae_model,z_mean,z_log_var
```

```python
vae,z_mean,z_log_var=modified_variational_autoencoder(input_img,n_filters=16,dropout=0.1,batchnorm=True)
# Compute VAE loss
def my_vae_loss(y_true, y_pred):
    xent_loss = 256*192*(keras.losses.mse(K.flatten(y_true), K.flatten(y_pred)))
    kl_loss = - 0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=1)
    vae_loss = K.mean(xent_loss + kl_loss)
    return vae_loss

vae.compile(optimizer='rmsprop', loss=my_vae_loss)
```
